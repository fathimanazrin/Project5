```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv("P5 FDI data.csv")
year_columns=['2000-2001','2001-2002','2002-2003','2003-2004','2004-2005','2005-2006','2006-2007','2007-2008','2008-2009','2009-2010','2010-2011','2011-2012','2012-2013','2013-2014','2014-2015','2015-2016','2016-2017']
data['Total_FDI'] = data[year_columns].sum(axis=1)
totalfdi=data.groupby('Sector')['Total_FDI'].sum().reset_index()
print(totalfdi)
sector=totalfdi.head(5)
label=sector['Sector']
value=sector['Total_FDI']
fig=plt.subplots(figsize=(10,8))
plt.title("Total FDI for Each Sector",fontsize=20)
plt.xlabel("Sector",fontsize=16)
plt.xticks(rotation=90)
plt.ylabel("Total FDI", fontsize=16)
plt.bar(label,value,label="Total FDI for Each Sector",color=('red','blue','yellow','green','pink'))
plt.show()
```

```python
Summary: Highlights which sectors received the most FDI, with the telecommunications and automobile industries dominating.
```

```python
total_fdi_per_year = data[year_columns].sum()
print(total_fdi_per_year)
plt.figure(figsize=(11, 6))
plt.bar(year_columns,total_fdi_per_year, color='skyblue')
plt.title('Total FDI Inflows per Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total FDI (in billion)', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
Summary:Shows how FDI inflows have surged, especially after 2005, indicating economic growth and increased international investment.
```

```python
automobile_data = data[data['Sector'] == 'AUTOMOBILE INDUSTRY']
year_columns = ['2000-2001', '2001-2002', '2002-2003', '2003-2004', 
                '2004-2005', '2005-2006', '2006-2007', '2007-2008', 
                '2008-2009', '2009-2010', '2010-2011', '2011-2012', 
                '2012-2013', '2013-2014', '2014-2015', '2015-2016', 
                '2016-2017']
automobile_data_long = pd.melt(automobile_data, id_vars='Sector', value_vars=year_columns, 
                               var_name='Year', value_name='FDI')
print(automobile_data_long)
label=automobile_data_long["FDI"]
value=automobile_data_long["Year"]
fig=plt.subplots(figsize=(18,7))
plt.title("Compare FDI trends across Automobile sectors over times",fontsize=20)
plt.plot(value, label,linewidth=3.0)
plt.xlabel("Years",fontsize=16)
plt.yticks(rotation=45)
plt.xticks(rotation=45)
plt.ylabel("Automobile Sector",fontsize=16)
plt.legend()
plt.show()
```

```python
Summary:This chart highlights the growth trajectory in the automobile industry and how it has attracted increasing amounts of FD
```
